```python
#library yg di guanakan
import pandas as pd
import numpy  as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression

#membaca file csv
df = pd.read_csv('TGM 2020-2023_eng.csv', sep = ";")

#eksplorasi data
df.info()
df.head()
```

```python
#pembersihan dan konversi data ke float
#fungsi konversi
def to_float(val):
    val = str(val).replace(',','.')
    return float(val) if val != 'N/A' else None

df['Internet Access Frequency per Week'] = df['Internet Access Frequency per Week'].apply(to_float)
df['Tingkat Kegemaran Membaca (Reading Interest)'] = df['Tingkat Kegemaran Membaca (Reading Interest)'].apply(to_float)

#pembersihan data
df_clean = df.dropna(subset = ['Internet Access Frequency per Week','Tingkat Kegemaran Membaca (Reading Interest)'])
```

```python
#statistik sederhana
stats = df_clean[['Internet Access Frequency per Week','Tingkat Kegemaran Membaca (Reading Interest)']].describe()
#mencari median
median_access = df_clean['Internet Access Frequency per Week'].median()
median_reading = df_clean['Tingkat Kegemaran Membaca (Reading Interest)'].median()
#mencari standar deviasi
std_access = df_clean['Internet Access Frequency per Week'].std()
std_reading = df_clean['Tingkat Kegemaran Membaca (Reading Interest)'].std()

print("======statistik sederhana=======")
print(stats)
print(f"\nmedian internet access : {median_access}")
print(f"median tingkat kegemearan membaca : {median_reading}")

print(f"\nstandar deviasi internet access : {std_access}")
print(f"standar deviasi kegemearan membaca : {std_reading}")
```

```python
#regresi linear sederhana
X = df_clean[['Internet Access Frequency per Week']]
y = df_clean['Tingkat Kegemaran Membaca (Reading Interest)']

model = LinearRegression()
model.fit(X, y)

print("intercept : ", model.intercept_)
print("koefisen : ",model.coef_[0])
print("R-squared : ",model.score(X,y))
```

```python
#visualisasi data
plt.figure (figsize=(8,6))
plt.scatter(X,y,color='blue',label = 'data observasi')
plt.plot(X,model.predict(X),color= 'black',label = 'garis regresi')
plt.title('regresi linear : akses internet vs minat baca di indonesia')
plt.xlabel('frekuensi akses internet perminggu')
plt.ylabel('Tingkat Kegemaran Membaca (Reading Interest)')
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()
```
